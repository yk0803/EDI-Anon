Copy Age. For All Folders.

```python
import os
import shutil

# Set the source and destination directories
source_dir = 'F:/Anonymized/EDI/Pix/Face_Recognition/'
dest_dir = 'F:/Anonymized/EDI/Pix/Age_Classification/'


# Get the subfolder names in the Old and Young directories
info_old = [f.name for f in os.scandir('F:/For_Info/Age_Class/Old') if f.is_dir()]
info_young = [f.name for f in os.scandir('F:/For_Info/Age_Class/Young') if f.is_dir()]

# Copy the relevant files to the destination directory
file_counter_old = 0
file_counter_young = 0

for iteration_dir in os.scandir(source_dir):
    if iteration_dir.is_dir():
        iteration_name = iteration_dir.name
        iteration_dest_dir = os.path.join(dest_dir, iteration_name)
        os.makedirs(iteration_dest_dir, exist_ok=True)

        subdir_old = os.path.join(iteration_dest_dir, 'Old')
        subdir_young = os.path.join(iteration_dest_dir, 'Young')
        os.makedirs(subdir_old, exist_ok=True)
        os.makedirs(subdir_young, exist_ok=True)

        for class_dir in os.scandir(iteration_dir.path):
            if class_dir.is_dir():
                class_name = class_dir.name

                if class_name in info_old:
                    class_dest_dir = subdir_old
                    file_counter = file_counter_old
                    file_counter_old += 1
                elif class_name in info_young:
                    class_dest_dir = subdir_young
                    file_counter = file_counter_young
                    file_counter_young += 1
                else:
                    continue

                os.makedirs(class_dest_dir, exist_ok=True)

                file_index = 0  # Initialize the file index

                for item in os.listdir(class_dir.path):
                    item_src = os.path.join(class_dir.path, item)
                    item_dest = os.path.join(class_dest_dir, f"{file_index}.jpg")
                    
                    while os.path.exists(item_dest):
                        # If the file already exists, increment the file index
                        file_index += 1
                        item_dest = os.path.join(class_dest_dir, f"{file_index}.jpg")
                    
                    shutil.copy2(item_src, item_dest)
                    file_index += 1  # Increment the file index for the next file
```

```python
import os
import shutil

# Source directories
source_dirs = [
    'D:/Younas_Work/D2_Final/Original/All FR/test',
    'D:/Younas_Work/D2_Final/Original/All FR/val'
]

# Target directory
target_dir = 'D:/Younas_Work/D2_Final/Original/FR the original'

# Iterate through each source directory
for source_dir in source_dirs:
    # Iterate through each folder in the source directory
    for folder_name in os.listdir(source_dir):
        source_folder = os.path.join(source_dir, folder_name)
        target_folder = os.path.join(target_dir, folder_name)
        
        # Check if the source folder exists
        if os.path.isdir(source_folder):
            # Create the target folder if it doesn't exist
            os.makedirs(target_folder, exist_ok=True)
            
            # Iterate through files in the source folder
            for filename in os.listdir(source_folder):
                source_file = os.path.join(source_folder, filename)
                target_file = os.path.join(target_folder, filename)
                
                # Check if the file already exists in the target folder
                if not os.path.exists(target_file):
                    # Copy the file from source to target
                    shutil.copy2(source_file, target_file)
```
