```python
# Define the evaluation function
def evaluate_model(loader, dataset_name):
    y_true = []
    y_scores = []
    test_loss = 0.0
    num_samples = 0

    model.eval()  # Set the model to evaluation mode
    with torch.no_grad():
        for inputs, labels in tqdm(loader):
            inputs = inputs.to(device)
            labels = labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)

            test_loss += loss.item() * inputs.size(0)
            num_samples += inputs.size(0)
            y_true += labels.cpu().tolist()
#             y_scores += outputs.cpu().tolist()
            
            softmax_probs = torch.softmax(outputs, dim=1)  # Apply softmax to convert logits to probabilities
            y_scores += softmax_probs.cpu().numpy().tolist()

    test_loss /= num_samples
    y_true = np.array(y_true)
    y_scores = np.array(y_scores)

    y_pred = np.argmax(y_scores, axis=1)
    test_acc = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred, average='macro')
    precision = precision_score(y_true, y_pred, average='macro')
    recall = recall_score(y_true, y_pred, average='macro')
    conf_mat = confusion_matrix(y_true, y_pred)

    # Calculate ROC curve and AUC for each class
#     fpr = dict()
#     tpr = dict()
#     roc_auc = dict()
#     for i in range(num_classes):
#         fpr[i], tpr[i], _ = roc_curve(y_true == i, y_scores[:, i])
#         roc_auc[i] = auc(fpr[i], tpr[i])

#     # Save the ROC plots for each class
#     for i in range(num_classes):
#         save_roc_plot(fpr[i], tpr[i], roc_auc[i], dataset_name, roc_save_dir, class_name=i)
        
#     roc_auc_macro = roc_auc_score(y_true, y_scores, average='macro')
    roc_auc_macro = roc_auc_score(y_true, y_scores, average='macro', multi_class='ovo')

    class_auc_scores = []

    # Iterate over each class
    for class_index in range(num_classes):
        # Create binary labels for the current class (1 for the current class, 0 for the rest)
        binary_labels = [1 if label == class_index else 0 for label in y_true]

        # Calculate AUC score for the current class
        auc_score = roc_auc_score(binary_labels, y_scores[:, class_index])  # predicted_labels[:, class_index] contains predicted labels for the current class

        # Append the AUC score to the list
        class_auc_scores.append(auc_score)

    # Calculate the macro-average AUC score
    macro_average_auc = sum(class_auc_scores) / num_classes  # num_classes is the number of classes


    return test_acc, f1, precision, recall, macro_average_auc
```

```python
# Create a combined result DataFrame for all datasets
combined_result_df = pd.DataFrame(columns=['Dataset', 'Accuracy', 'F1 Score', 'Precision', 'Recall', 'AUC'])


# Evaluation loop
for dataset_dir in dataset_dirs:
    test_dataset = ImageFolder(root=dataset_dir, transform=transform)
    test_loader = DataLoader(test_dataset, batch_size=1, shuffle=True, pin_memory=True)

#     dataset_name = os.path.basename(dataset_dir)  
    # Extract the dataset name from the directory
    
    path_components = os.path.normpath(dataset_dir).split(os.path.sep)
    dataset_name = os.path.join(*path_components[-3:])
    dataset_name = dataset_name.replace(os.path.sep, '_')

    test_acc, f1, precision, recall, roc_auc_macro = evaluate_model(test_loader, dataset_name)
    
    print(f'Dataset: {dataset_name}')
    print(f'Accuracy: {test_acc}')
    print(f'F1 Score: {f1}')
    print(f'Precision: {precision}')
    print(f'Recall: {recall}')
    print(f'AUC: {roc_auc_macro}')

    # Append the dataset results to the combined result DataFrame
    combined_result_df = combined_result_df.append({'Dataset': dataset_name,
                                                    'Test Accuracy': test_acc,
                                                    'F1 Score': f1,
                                                    'Precision': precision,
                                                    'Recall': recall,
                                                    'AUC': roc_auc_macro},
                                                   ignore_index=True)

# Multiply the accuracy values by 100 and round to the second digit after the decimal point
combined_result_df['Test Accuracy'] = (combined_result_df['Test Accuracy'] * 100).round(2)

# Save the combined result DataFrame to an Excel file with values rounded to the second digit
if not os.path.exists(excel_file_path):
    combined_result_df.to_excel(excel_file_path, index=False)
else:
    with pd.ExcelWriter(excel_file_path, mode='a', engine='openpyxl') as writer:
        combined_result_df.to_excel(writer, index=False, header=False, sheet_name='Results')

# Print message after processing all datasets
print('Evaluation complete. Combined results saved in:', excel_file_path)
```
