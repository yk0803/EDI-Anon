```python
def save_roc_plot(fpr, tpr, auc, dataset_dir, ROC_dir):
    path_components = os.path.normpath(dataset_dir).split(os.path.sep)
    dataset_name = os.path.join(*path_components[-3:])
    dataset_name = dataset_name.replace(os.path.sep, '_')
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve')
    plt.legend(loc='lower right')
    plt.savefig(os.path.join(ROC_dir, f'ROC_{dataset_name}.png'))
    plt.close()
```

```python
# Create a combined result DataFrame for all datasets
combined_result_df = pd.DataFrame(columns=['Dataset', 'Balanced Accuracy', 'Accuracy', 'F1 Score', 'Precision', 'Recall', 'AUC'])

model.eval()

# Iterate over each dataset directory
for dataset_dir in dataset_dirs:
    # Load the dataset
    dataset = ImageFolder(root=dataset_dir, transform=transform)
    test_loader = DataLoader(dataset, batch_size=batch_size)

    predicted_labels = []
    true_labels = []

    with torch.no_grad():
        for inputs, labels in tqdm(test_loader):
            # Move inputs and labels to the appropriate device
            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = model(inputs)
            _, predicted = torch.max(outputs.data, 1)
            predicted_labels.extend(predicted.cpu().tolist())
            true_labels.extend(labels.cpu().tolist())
        
    # Calculate balanced accuracy
    class_labels = np.unique(true_labels)
    class_correct_val = [np.sum(np.logical_and(predicted_labels == label, true_labels == label)) for label in class_labels]
    class_total_val = [np.sum(true_labels == label) for label in class_labels]

    class_correct_val = np.array(class_correct_val)
    class_total_val = np.array(class_total_val)

    # Calculate the class weights based on the inverse of the class frequencies
    class_weights = 1 / (class_total_val + 1e-8)  # Adding a small epsilon to avoid division by zero

    # Calculate the weighted sum of correct predictions for each class
    weighted_correct_sum = np.sum(class_correct_val * class_weights)

    # Calculate the weighted sum of total samples for each class
    weighted_total_sum = np.sum(class_total_val * class_weights)

    # Calculate the balanced accuracy using the weighted sums
    balanced_accuracy = weighted_correct_sum / weighted_total_sum

    # Multiply balanced accuracy by 100 and round to the second decimal digit
    balanced_accuracy = round(balanced_accuracy * 100, 2)


    # Calculate other metrics (Accuracy, F1 score, Precision, Recall, AUC)
    accuracy = accuracy_score(true_labels, predicted_labels)
    f1 = f1_score(true_labels, predicted_labels, average='weighted')
    precision = precision_score(true_labels, predicted_labels, average='weighted')
    recall = recall_score(true_labels, predicted_labels, average='weighted')
    
    # Plot and save the ROC curve for the current dataset    
    fpr, tpr, _ = roc_curve(true_labels, predicted_labels)
    auc = roc_auc_score(true_labels, predicted_labels)
    save_roc_plot(fpr, tpr, auc, dataset_dir, ROC_dir)

    print(f'Dataset: {dataset_dir}')
    print(f'Balanced Accuracy: {balanced_accuracy}')
    print(f'Accuracy: {accuracy}')
    print(f'F1 Score: {f1}')
    print(f'Precision: {precision}')
    print(f'Recall: {recall}')
    
    # Append the dataset results to the combined result DataFrame
    combined_result_df = combined_result_df.append({'Dataset': dataset_dir,
                                                    'Balanced Accuracy': balanced_accuracy,
                                                    'Accuracy': accuracy,
                                                    'F1 Score': f1,
                                                    'Precision': precision,
                                                    'Recall': recall,
                                                    'AUC': auc},
                                                   ignore_index=True)


# Save the combined result DataFrame to an Excel file
combined_result_file_path = os.path.join(excel_file_path, 'D:/Younas_Work/D2_Final/Results_Final/Excel_Sheets/Eval_Age_All.xlsx')
if not os.path.exists(combined_result_file_path):
    combined_result_df.to_excel(combined_result_file_path, index=False)

# Otherwise, append to the existing Excel file
else:
    with pd.ExcelWriter(combined_result_file_path, mode='a', engine='openpyxl') as writer:
        combined_result_df.to_excel(writer, index=False, header=False, sheet_name='Results')

# Print message after processing all datasets
print('Evaluation complete. Combined results saved in:', combined_result_file_path)
```
